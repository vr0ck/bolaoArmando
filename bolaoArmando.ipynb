```python
import random
numeros = list(range(1,61))
#print(numeros)

jogos = []
for i in range(10):
    jogo = []
    for n in range(6):
        atual = random.randrange(len(numeros))
        #print('atual',atual,numeros[atual])
        jogo.append(numeros[atual])
        numeros.pop(atual)
        jogo.sort()
    jogos.append(jogo)
    #print('numeros',numeros)
    #print('jogos',jogos)
for jogo in jogos:
    print("jogo: [", "%02d" % (jogos.index(jogo) + 1), ']', end=' ')
    for dez in jogo:
        print("%02d" % dez, end=' ')
    print('\n')
  
```
